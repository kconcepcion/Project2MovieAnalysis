```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#reading in the csv file
movieList = 'movieFiles1.csv'
movieDf = pd.read_csv(movieList, low_memory=False)

movieDf
```

```python
#dropping duplicates
movieDf = movieDf.drop_duplicates(subset=['title'])
```

```python
#selecting only english movies
movieDf = movieDf.loc[movieDf["original_language"] == 'en']
movieDf
```

```python
#dropping movies that have not premiered
movieDf = movieDf.loc[movieDf["vote_count"] != 0]
movieDf
```

```python
movieDf.drop(columns=['original_title', 'vote_count'])
movieDf
```

```python
movieDf.to_csv('movies1.csv')
```
