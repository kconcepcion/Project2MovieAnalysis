```python
#Import Dependencies
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import numpy as np
#%matplotlib inline


#inport line from where beautiful soup will be pulling data ("Main Link")
url = 'https://www.boxofficemojo.com/year/2021/?ref_=bo_yl_table_1'
html_text = requests.get(url).text
soup = BeautifulSoup(html_text, 'html.parser')
```

<h1>Make data into list</h1>
<h4>list consist of (rank, title, gross, release, studio)</h4>

```python
#Read html where titles of movies are located 
titles=soup.find_all('td', class_='a-text-left mojo-field-type-release mojo-cell-wide')


#for loop to look thru all the html titles and convert them to string
All_titles = []
for thing in titles:
    All_titles.append(thing.select('a')[0].string)
```

```python
#print all titles 
All_titles
```

```python
#Read html where studio of movies are located 
studio=soup.find_all('td', class_='a-text-left mojo-field-type-studio')
   
                   
                     
#for loop to look thru all the html studio and convert them to string                     
All_studio = []
for thing in studio:
    All_studio.append(thing.text[:-2])                   
```

```python
#print all studio 
All_studio
```

```python
#Read html where rank of movies are located 
rank=soup.find_all('td', class_='a-text-right mojo-header-column mojo-truncate mojo-field-type-rank mojo-sort-column')



#for loop to look thru all the html rank and convert them to string      
All_rank = []
for thing in rank:
    All_rank.append(thing.string)

```

```python
#print all rank 
All_rank
```

```python
#Read html where rank of gross1 are located 
gross=soup.find_all('td', class_='a-text-right mojo-field-type-money mojo-estimatable')



#for loop to look thru all the html gross1 and convert them to string    
All_gross = []
for thing in gross:
    All_gross.append(thing.string)
```

```python
#print all gross 
All_gross
```

```python
#Read html where release of release are located 
release=soup.find_all('td', class_='a-text-left mojo-field-type-date a-nowrap')



#for loop to look thru all the html release and convert them to string 
All_release = []
for thing in release:
    All_release.append(thing.string)
```

```python
#print all release 
All_release
```

<h1>Converting List Into Dataframe</h1>

```python
#Create list "All_titles, All_studio, All_rank, All_release"
df = pd.DataFrame(
{
    "rank": All_rank,
    "title": All_titles,
    "studio": All_studio,
    "release": All_release,
})
df.head()
```

```python
#Create list "All_gross1" had to make into different dataframe
#due to inconsitency with length of data
df1 = pd.DataFrame(
{

    "gross": All_gross,

})
df1.head()
```

```python
df
```

```python
#drop every other 2nd row of df1 to correct duplicate gross totals found in df1
df1 = df1.iloc[::2]
```

```python
#reset index for df1
df1 = df1.reset_index()
```

<h1>Combining dataframes</h1>

```python
#combine dataframes df to df1 
df = df.join(df1)
```

```python
#convert files to csv
df.to_csv
```
